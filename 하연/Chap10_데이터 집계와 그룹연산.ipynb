### 10.3 apply 메서드 : 일반적인 분리 적용 병합

```python
#특정 열에서 가장 큰 값을 갖는 행을 선택하는 함수
def top(df, n=5, column = "tip_pct"):
    return df.sort_values(column, ascending=False)[:n]
```

#### 10.3.2 사분위수 분석과 버킷분석

```python
frame = pd.DataFrame({"data1":np.random.standard_normal(1000), "data2":np.random.standard_normal(1000)})
frame.head()
```

```python
quartiles = pd.cut(frame["data1"],4)
quartiles.head(10)
```

```python
#cut 에서 반환된 Categorical 객체는 바로 groupby로 넘길 수 있다
#quartiles에 대한 그룹 통계는 다음과 같이 계산
def get_stats(group):
    return pd.DataFrame(
    {"min":group.min(), "max":group.max(), "count":group.count(),"mean":group.mean()})
```

```python
grouped = frame.groupby(quartiles)
grouped.apply(get_stats)
```

```python
grouped.agg(["min","max","count","mean"])
```

```python
#샘플을 사분위수로 나누기 위해 버킷 개수로 4를 넘기고 labels=False를 전달해 간격 대신 사분위수 색인을 구한다 (?)
quartiles_samp = pd.qcut(frame["data1"],4,labels=False)
quartiles_samp.head()
```

```python
grouped = frame.groupby(quartiles_samp)
grouped.apply(get_stats)
```

#### 10.3.3 그룹별 값으로 결측치 채우기

```python
#누락된 값을 평균값으로 대체하자
s = pd.Series(np.random.standard_normal(6))
s[::2]= np.nan
```

```python
#그룹별로 채워넣고 싶은 값이 다르다면
#1. 데이터를 그룹으로 나누고 apply 함수를 이용해 각 그룹에 fillna를 적용한다
states = ["Ohio","New York","Vermont","Florida","Oregon","Nevada","California","Idaho"]
group_key = ["East","East","East","East","West","West","West","West",]
data = pd.Series(np.random.standard_normal(8), index = states)
```

```python
data
```

```python
#몇몇을 결측치로 만든다
data[["Vermont","Nevada","Idaho"]] = np.nan
```

```python
data.groupby(group_key).size()
```

```python
data.groupby(group_key).count()
```

```python
data.groupby(group_key).mean()
```

```python
#다음과 같이 그룹의 평균값으로 누락된 값을 채울 수 있음
def fill_mean(group):
    return group.fillna(group.mean())
data.groupby(group_key).apply(fill_mean)
```

```python
#2. 그룹별로 미리 정의된 다른 값 채워넣기
fill_values={"East":0.5, "West":-1}
def fill_func(group):
    return group.fillna(fill_values[group.name])

data.groupby(group_key).apply(fill_func)
```

#### 10.3.4 랜덤 표본과 순열

```python
#Series의 sample 메서드를 사용
suits = ["H","S","C","D"] #하트, 스페이드, 클럽, 다이아몬드
card_val = (list(range(1,11)) + [10] * 3)*4
base_names = ["A"]+list(range(2,11)) + ["J","K","Q"]
cards = []
for suit in suits :
    cards.extend(str(num) + suit for num in base_names)
    
deck = pd.Series(card_val, index=cards)
```

```python
deck.head(13)
```

```python
#다섯장의 카드를 뽑는가
def draw(deck, n=5):
    return deck.sample(n)
draw(deck)
```

```python
def get_suit(card):
    #마지막 글자는 모양 표시
    return card[-1]
```

```python
#무작위로 뽑자
deck.groupby(get_suit).apply(draw,n=2)
```

```python
deck.groupby(get_suit, group_keys=False).apply(draw, n=2)
```

#### 10.3.5 그룹 가중평균과 상관관계

```python
df = pd.DataFrame({"category":["a","a","a","a","b","b","b","b"], "data":np.random.standard_normal(8), "weights":np.random.uniform(size=8)})
```

```python
df
```

```python
#그룹별 가중평균
grouped = df.groupby("category")
def get_wavg(group):
    return np.average(group["data"],weights=group["weights"])
grouped.apply(get_wavg)
```

```python
close_px = pd.read_csv("examples/stock_px.csv",parse_dates=True, index_col=0)
```

```python
close_px.info()
```

```python
close_px.tail(4)
```

```python
#"SPX"열과 다른 열의 상관관계를 계산하자
def spx_corr(group):
    return group.corrwith(group["SPX"])
```

```python
#pct_change 함수를 이용해서 close_px의 퍼센트 변화율을 계산한다
rets = close_px.pct_change().dropna()
```

```python
#연도별 퍼센트 변화율을 구한다
def get_year(x):
    return x.year
by_year = rets.groupby(get_year)
by_year.apply(spx_corr)
```

```python
#두 열의 상관관계를 계산한다
def corr_aapl_msft(group):
    return group["AAPL"].corr(group["MSFT"])
by_year.apply(corr_aapl_msft)
```

#### 10.3.6 그룹별 선형회귀

```python
#회귀함수를 작성하고 데이터 묶음마다 최소제곱법으로 회귀
import statsmodels.api as sm
def regress(data, yvar=None, xvars=None):
    Y = data[yvar]
    X = data[xvars]
    X["intercept"] = 1.
    result = sm.OLS(Y,X).fit()
    return result.params
```

```python
# SPX 수익률에 대한 애플 주식의 연간 선형 회귀
by_year.apply(regress, yvar="AAPL",xvars=["SPX"])
```

## 10.4 그룹 변환과 래핑되지 않은 groupby

- transform메서드는 apply와 유사하지만 더 많은 제약을 갖는다
- 그룹의 모양대로 브로드캐스팅할 스칼라 값을 생성할 수 있음
- 입력 그룹과 동일한 모양의 개체를 생성할 수 있음
- 입력 변경 X

```python
df = pd.DataFrame({'key':['a','b','c']*4, 'value':np.arange(12.)})
```

```python
df
```

```python
#키별 그룹평균
g= df.groupby("key")['value']
```

```python
g.mean()
```

```python
#transform으로 단일 그룹의 평균을 계산
def get_mean(group):
    return group.mean()
```

```python
g.transform(get_mean)
```

```python
g.transform('mean')
```

```python
def times_two(group):
    return group*2
```

```python
#transform은 입력과 크기가 같아야함
g.transform(times_two)
```

```python
def get_ranks(group):
    #그룹의 순위를 내림차순으로 계산
    return group.rank(ascending = False)
g.transform(get_ranks)
```

```python
#단순 집계로 구성된 그룹 변환 함수
def normalize(x):
    return (x-x.mean())/x.std()
```

```python
g.transform(normalize)
```

```python
g.apply(normalize)
```
